```python
import json

def json_file_to_dict(file_path):
    try:
        with open(file_path, 'r') as file:
            data = json.load(file)
            return data
    except FileNotFoundError:
        print("Error: File not found.")
        return {}
    except json.JSONDecodeError:
        print("Error: Invalid JSON data.")
        return {}

# Replace 'path/to/your/file.json' with the actual path to your JSON file
json_file_path = 'mem_event_data.json'
resulting_dict = json_file_to_dict(json_file_path)
```

```python
ct=0
print(len(resulting_dict))
for i in resulting_dict:
    print(i,resulting_dict[i])
    ct+=1
    if ct>4:
        break
```

## convert the mem_event to event_mem file for adj list

```python
input_dict = resulting_dict

output_dict = {}

for member, events in input_dict.items():
    for event in events:
        if event not in output_dict:
            output_dict[event] = [member]
        else:
            output_dict[event].append(member)

print(len(output_dict))
```

```python
ct=0
print(len(output_dict))
for i in output_dict:
    print(i,output_dict[i])
    ct+=1
    if ct>4:
        break
```

```python
s=set(output_dict.keys())
m=set(resulting_dict.keys())
len(s.intersection(m))
```

```python
fresh=s^m
unique_dic={}
for i in fresh:
    unique_dic[i]=output_dict[i] if i in s else resulting_dict[i]
```

```python
len(unique_dic)
```

```python
ct=0
print(len(unique_dic))
for i in unique_dic:
    print(i,unique_dic[i])
    ct+=1
    if ct>4:
        break
```

```python
# Specify the file path where you want to save the JSON data
import json
file_path = 'event_mem_adjacency_data.json'##create json file for event
with open(file_path, 'w') as json_file:
    json.dump(unique_dic, json_file)

print("Data saved to JSON file:", file_path)
```
